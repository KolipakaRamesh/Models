```python
import pandas as pd
import random

# Create a list of 5 cities (or more, if you prefer)
cities = ["New York", "London", "Paris", "Tokyo", "Sydney"]

# Generate sample weather data for each city with the specified features
data = {
    "City": cities,
    "Temperature (°C)": [round(random.uniform(-10, 35), 1) for _ in cities],
    "Humidity (%)": [random.randint(30, 100) for _ in cities],
    "Pressure (hPa)": [round(random.uniform(980, 1030), 1) for _ in cities],
    "Wind Speed (m/s)": [round(random.uniform(0, 15), 1) for _ in cities],
    # Add a target variable, for example, predicting Temperature based on other features
    # or another relevant weather metric if you have one in mind.
    # For this example, let's assume we are training a model to predict Temperature.
}

# Create a DataFrame
weather_df_new = pd.DataFrame(data)

# Display the DataFrame
display(weather_df_new)
```

```python
# Assuming 'weather_df_new' is the DataFrame created in the previous step

# Define features (X) and target (y)
# We'll use Humidity, Pressure, and Wind Speed to predict Temperature
X_new = weather_df_new[['Humidity (%)', 'Pressure (hPa)', 'Wind Speed (m/s)']]
y_new = weather_df_new['Temperature (°C)']

# Train a model (using RandomForestRegressor again as an example)
from sklearn.ensemble import RandomForestRegressor
model_new = RandomForestRegressor()
model_new.fit(X_new, y_new)

# Save the new model
import joblib
joblib.dump(model_new, 'weather_model.pkl')

# Download the new model file
from google.colab import files
files.download('weather_model.pkl')
```
